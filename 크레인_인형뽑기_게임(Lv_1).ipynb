## 첫 번째 방법
1. moves에서 원하는 곳이 0이 아닐 때 board를 순차적으로 탐색하는 이중 for문 생성.
2. 0이 아니면 만족하는 board의 제일 앞에 인덱스에 있는 원소를 basket에 append하고 board의 원소를 0으로 변환.
3. basket의 크기가 1보다 크면 basket의 맨 뒷부분의 앞뒤 원소를 비교해 같으면 pop함수를 사용해 원소를 꺼내고 answer을 2씩 카운트 후 answer을 return.

결과 : 성공

```python
def solution(board, moves):
    answer = 0
    basket = []
    for i in moves:
        for j in range(len(board)):
            if board[j][i - 1] != 0:
                basket.append(board[j][i - 1])
                board[j][i - 1] = 0
                
                if len(basket) > 1:
                    if basket[len(basket) - 2] == basket[len(basket) - 1]:
                        basket.pop()
                        basket.pop()
                        answer += 2
                break   
    return answer
solution([[0,0,0,0,0],[0,0,1,0,3],[0,2,5,0,1],[4,2,4,4,2],[3,5,1,3,1]], [1,5,3,5,1,2,1,4])
```
