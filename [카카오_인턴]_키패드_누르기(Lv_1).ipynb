## 문제 설명
스마트폰 전화 키패드의 각 칸에 다음과 같이 숫자들이 적혀 있습니다.

image.png

이 전화 키패드에서 왼손과 오른손의 엄지손가락만을 이용해서 숫자만을 입력하려고 합니다.
맨 처음 왼손 엄지손가락은 * 키패드에 오른손 엄지손가락은 # 키패드 위치에서 시작하며, 엄지손가락을 사용하는 규칙은 다음과 같습니다.

1. 엄지손가락은 상하좌우 4가지 방향으로만 이동할 수 있으며 키패드 이동 한 칸은 거리로 1에 해당합니다.
2. 쪽 열의 3개의 숫자 1, 4, 7을 입력할 때는 왼손 엄지손가락을 사용합니다.
3. 오른쪽 열의 3개의 숫자 3, 6, 9를 입력할 때는 오른손 엄지손가락을 사용합니다.
4. 가운데 열의 4개의 숫자 2, 5, 8, 0을 입력할 때는 두 엄지손가락의 현재 키패드의 위치에서 더 가까운 엄지손가락을 사용합니다.

    4-1. 만약 두 엄지손가락의 거리가 같다면, 오른손잡이는 오른손 엄지손가락, 왼손잡이는 왼손 엄지손가락을 사용합니다.

순서대로 누를 번호가 담긴 배열 numbers, 왼손잡이인지 오른손잡이인 지를 나타내는 문자열 hand가 매개변수로 주어질 때, 각 번호를 누른 엄지손가락이 왼손인 지 오른손인 지를 나타내는 연속된 문자열 형태로 return 하도록 solution 함수를 완성해주세요.

## [제한사항]
- numbers 배열의 크기는 1 이상 1,000 이하입니다.
- numbers 배열 원소의 값은 0 이상 9 이하인 정수입니다.
- hand는 "left" 또는 "right" 입니다.
    - "left"는 왼손잡이, "right"는 오른손잡이를 의미합니다.
- 왼손 엄지손가락을 사용한 경우는 L, 오른손 엄지손가락을 사용한 경우는 R을 순서대로 이어붙여 문자열 형태로 return 해주세요.

## 첫 번째 방법
1. dis라는 배열을 만들고 키패드의 각 숫자를 2차 행렬로 만들었을 때 0 ~ 9의 인덱스를 저장.
2. numbers를 하나하나 탐색하는 반복하는 반복문으로 i가 1,4,7이면 answer에 'L'을 추가한 후 lcur에 i를 추가
3. i가 3, 6, 9이면 answer에 'R'을 추가한 후 rcur에 i를 추가.
4. i가 left, right 없는 수면 lcur과 rcur 사이에 거리를 구함(현재 탐색하고 있는 i와 최근 오른쪽 손가락, 왼쪽 손가락의 위치를 비교).
5. ldis가 rdis 보다 클 경우 answer에 'R'을 추가 ldis가 rdis보다 작을 경우 answer에 'L'을 추가.
6. 만약 거리가 같으면 입력받은 hand를 확인해 right면 'R'을, left면 'L'을 추가한 후 answer을 return.

## 두 번째 방법
첫 번째 방법 중 1번에서 키패드의 "*"와 "#"의 위치도 추가하여 처음 손가락의 위치를 "*", "#"에 즉, lcur, rcur 각각 10, 11을 초기 값으로 설정.

결과 : 성공

```python
def solution(numbers, hand):
    answer = ''
    lcur, rcur = 10, 11
    left = [1, 4, 7]
    right = [3, 6, 9]
    dis = [[3, 1], [0, 0], [0, 1], [0, 2], [1, 0], [1, 1], [1, 2], [2, 0], [2, 1], [2, 2], [3, 0], [3, 2]]

    for i in numbers:
        print(i)
        if i in left:
            answer += 'L'
            lcur = i
        elif i in right:
            answer += 'R'
            rcur = i
        else:
            ldis = abs(dis[i][0] - dis[lcur][0]) + abs(dis[i][1] - dis[lcur][1])
            rdis = abs(dis[i][0] - dis[rcur][0]) + abs(dis[i][1] - dis[rcur][1])
            if ldis > rdis:
                answer += 'R'
                rcur = i
            elif ldis < rdis:
                answer += 'L'
                lcur = i
            else:
                if hand =='right':
                    answer += 'R'
                    rcur = i
                else:
                    answer += 'L'
                    lcur = i
    return answer

solution([7, 0, 8, 2, 8, 3, 1, 5, 7, 6, 2], "left")
```
