위와 같은 삼각형의 꼭대기에서 바닥까지 이어지는 경로 중, 거쳐간 숫자의 합이 가장 큰 경우를 찾아보려고 합니다. 아래 칸으로 이동할 때는 대각선 방향으로 한 칸 오른쪽 또는 왼쪽으로만 이동 가능합니다. 예를 들어 3에서는 그 아래칸의 8 또는 1로만 이동이 가능합니다.

삼각형의 정보가 담긴 배열 triangle이 매개변수로 주어질 때, 거쳐간 숫자의 최댓값을 return 하도록 solution 함수를 완성하세요.

## 제한사항

- 삼각형의 높이는 1 이상 500 이하입니다.
- 삼각형을 이루고 있는 숫자는 0 이상 9,999 이하의 정수입니다.

## 첫 번째 방법

1. 입력받은 triangle의 첫 번째와 두 번째 원소를 memo에 저장하고 그 외 나머지 memo에 triangle과 같은 크기의 행렬 0을로 초기화.
2. memo의 모든 원소를 순회하면서 거쳐간 수의 합을 저장  단, 양 끝 원소가 아닌 중간 원소는 거쳐간 수의 합이 두 개씩 존재하므로 둘 줄 더 큰 값으로 저장.
3. memo의 마지막 배열에서 가장 큰 값을 return.

결과 : 성공


```python
def solution(triangle):
    memo = [triangle[0], triangle[1]]

    for i in range(3, len(triangle) + 1):
        memo.append([0] * i)

    for i in range(1, len(memo)):
        for j in range(len(memo[i])):
            if j == 0:
                if i == 1:
                    memo[i][j] = triangle[i][j] + memo[0][0]
                else:
                    memo[i][j] = triangle[i][j] + memo[i - 1][0]
            elif j == len(memo[i]) - 1:
                if i == 1:
                    memo[i][j] = triangle[i][j] + memo[0][0]
                else:
                    memo[i][j] = triangle[i][j] + memo[i - 1][j - 1]
            else:
                memo[i][j] = max((triangle[i][j] + memo[i - 1][j]), (triangle[i][j] + memo[i - 1][j - 1]))
    return max(memo[-1])
solution([[7], [3, 8], [8, 1, 0], [2, 7, 4, 4], [4, 5, 2, 6, 5]])
```
