# 문제 - 완주하지 못한 선수
[문제 링크](https://school.programmers.co.kr/learn/courses/30/lessons/42576)

## 문제 설명 
수많은 마라톤 선수들이 마라톤에 참여하였습니다. 단 한 명의 선수를 제외하고는 모든 선수가 마라톤을 완주하였습니다.

마라톤에 참여한 선수들의 이름이 담긴 배열 participant와 완주한 선수들의 이름이 담긴 배열 completion이 주어질 때, 완주하지 못한 선수의 이름을 return 하도록 solution 함수를 작성해주세요.

## 제한 사항
- 마라톤 경기에 참여한 선수의 수는 1명 이상 100,000명 이하입니다.
- completion의 길이는 participant의 길이보다 1 작습니다. 
- 참가자의 이름은 1개 이상 20개 이하의 알파벳 소문자로 이루어져 있습니다.
- 참가자 중에는 동명이인이 있을 수 있습니다.

## 입출력 예

image.png

## 입출력 예 설명

###예제 #1
"leo"는 참여자 명단에는 있지만, 완주자 명단에는 없기 때문에 완주하지 못했습니다.

###예제 #2
"vinko"는 참여자 명단에는 있지만, 완주자 명단에는 없기 때문에 완주하지 못했습니다.

###예제 #3
"mislav"는 참여자 명단에는 두 명이 있지만, 완주자 명단에는 한 명밖에 없기 때문에 한명은 완주하지 못했습니다.

## 첫 번째 방법

1. completion를 i로 하나씩 탐색해 i가 participant에 있으면 i를 삭제.
2. participant를 문자열로 return.

결과 : 실패

-> 효율성 테스트 실패

```python
def solution(participant, completion):
    
    for i in completion:
        if i in participant:
            participant.remove(i)
    return ''.join(participant)

solution(["mislav", "stanko", "mislav", "ana"], ["stanko", "ana", "mislav"])
```

## 두 번째 방법

1. list를 사용하는 것보다 dictionary를 사용하는 것이 효율성 높이기 위해 더 효과적이기 때문에 participant의 원소를 키, 값은 0으로 초기화된 dic을 선언.
2. participant를 i로 탐색해 i를 키로 가지는 값을 카운트.
3. completion를 i로 탐색해 i를 키도 가지는 값을 디스카운트.
4. dic에서 값이 0이 아닌 수를 answer에 저장한 후 return.

결과 : 성공

P.S. 문제를 해결할 때 문제의 분류를 인지하고 풀면 문제를 보다 효율적으로 풀 수 있고 문제 해결 시간도 줄일 수 있음.

```python
def solution(participant, completion):
    answer = ''
    dic = {string : 0 for string in participant}

    for i in participant:
        dic[i] += 1
    
    for i in completion:
        dic[i] -= 1
    
    for key, value in dic.items():
        if value != 0:
            answer += key
    return answer
solution(["mislav", "stanko", "mislav", "ana"], ["stanko", "ana", "mislav"])
```
