## 문제 설명

다음 규칙을 지키는 문자열을 올바른 괄호 문자열이라고 정의합니다.

- (), [], {} 는 모두 올바른 괄호 문자열입니다.
- 만약 A가 올바른 괄호 문자열이라면, (A), [A], {A} 도 올바른 괄호 문자열입니다. 예를 들어, [] 가 올바른 괄호 문자열이므로, ([]) 도 올바른 괄호 문자열입니다.
- 만약 A, B가 올바른 괄호 문자열이라면, AB 도 올바른 괄호 문자열입니다. 예를 들어, {} 와 ([]) 가 올바른 괄호 문자열이므로, {}([]) 도 올바른 괄호 문자열입니다.

대괄호, 중괄호, 그리고 소괄호로 이루어진 문자열 s가 매개변수로 주어집니다. 이 s를 왼쪽으로 x (0 ≤ x < (s의 길이)) 칸만큼 회전시켰을 때 s가 올바른 괄호 문자열이 되게 하는 x의 개수를 return 하도록 solution 함수를 완성해주세요.

## 제한 사항

s의 길이는 1 이상 1,000 이하입니다.

## 첫 번째 방법

1. stack을 활용하여 대응하는 괄호가 연달아 들어오면 stack에서 pop하고 stack이 비어있는지 비어있지 않은지 판별하는 check함수를 선언.
2. s의 크기만큼 반복하는 반복문으로 앞에서 선언한 check함수를 사용해 그 값이 True면 answer을 카운트.
3. s의 0번째 원소를 s의 마지막 원소로 두고 반복문이 종료되면 answer을 return.

결과 : 성공


```python
def check(s):
    stack = [s[0]]

    for i in range(1, len(s)):
        stack.append(s[i])
        while len(stack) >= 2:
            if stack[-1] == ")" and stack[-2] == "(":
                stack.pop()
                stack.pop()
            elif stack[-1] == "]" and stack[-2] == "[":
                stack.pop()
                stack.pop()
            elif stack[-1] == "}" and stack[-2] == "{":
                stack.pop()
                stack.pop()
            else:
                break
    if not stack:
        return True
    else:
        return False

def solution(s):
    answer = 0
    for _ in range(len(s)):
        if check(s):
            answer += 1
        tmp = s[0]
        s = s[1 : ] + tmp
    return answer

solution("[](){}")
```
