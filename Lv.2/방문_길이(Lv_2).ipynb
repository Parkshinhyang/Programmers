## 첫 번째 방법

1. 입력된 dirs를 i로 탐색해 dis_x, dis_y에서 i를 키로 하는 값을 before_x, before_y에 더해줌.
2. foot에 이동 전 위치와 이동 후 위치를 정렬해 튜플형태로 저장.(방향성을 따지지 않기 때문에 중복 제거의 편의성을 위한 정렬.)
3. check에 foot을 add해주고 cur을 이동 후 위치로 설정.
4. check의 크기를 return.

결과 : 성공

```python
def solution(dirs):
    dis_x = {"U" : 0, "D" : 0, "R" : 1, "L" : -1}
    dis_y = {"U" : 1, "D" : -1, "R" : 0, "L" : 0}

    cur = (0, 0)
    check = set()
    
    for i in dirs:
        before_x, before_y = cur
        after_x = before_x + dis_x[i]
        after_y = before_y + dis_y[i]

        if -5 <= after_x <= 5 and -5 <= after_y <= 5:
            foot = tuple(sorted([(before_x, before_y), (after_x, after_y)]))
            check.add(foot)
            cur = (after_x, after_y)
    return len(check)

solution("ULURRDLLU")
```
