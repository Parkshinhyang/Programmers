## 첫 번째 방법

1. 각 행의 수가 점진적으로 늘어나 마지막 행의 원소 수가 n개인 2차 행렬 triangle을 0으로 초기화.
2. 삼각형의 모양을 하기 때문에 모서리의 수인 3으로 나눈 나머지로 하, 우, 상 방향을 구함.
3. 각 방향에 맞게 x와 y를 더하고 빼 triangle의 인덱스에 맞는 값을 저장.
4. 2차 배열 triangle을 각 배열을 기준으로 더한 1차 배열 answer을 return.

결과 : 성공

```python
def solution(n):
    answer = []
    x, y, num = -1, 0, 1
    triangle = [[0] * i for i in range(1, n + 1)]
    
    for i in range(n):
        for _ in range(i, n):
            if i % 3 == 0:
                x += 1
            elif i % 3 == 1:
                y += 1
            else:
                x -= 1
                y -= 1
            triangle[x][y] = num
            num += 1
    return sum(triangle, [])
                
solution(4)
```
