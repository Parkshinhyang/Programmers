## 첫 번째 방법

1. 입력받은 n이 2, 3이면 각각 4, 11을 return.
2. n이 2, 3이 아니라면 4로 초기화된 num 2씩 더해가며 f(n) = 4 * f(n - 2) - f(n - 4) 점화식으로 [3, 11]으로 초기화된 check을 반복해서 초기화.
3. 주어진 조건에 맞게 check[1]을 1000000007로 나눈 나머지를 return.

결과 : 성공

```python
def solution(n):
    answer = 0

    if n == 2:
        return 3
    if n == 4:
        return 11
    check = [3, 11]
    num = 4

    while num < n:
        check = [check[1], check[1] * 4 - check[0]]
        num += 2
    return check[1] % 1000000007
solution(6)
```
