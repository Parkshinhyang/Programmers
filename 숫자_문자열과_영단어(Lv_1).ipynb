네오와 프로도가 숫자놀이를 하고 있습니다. 네오가 프로도에게 숫자를 건넬 때 일부 자릿수를 영단어로 바꾼 카드를 건네주면 프로도는 원래 숫자를 찾는 게임입니다.

다음은 숫자의 일부 자릿수를 영단어로 바꾸는 예시입니다.

- 1478 → "one4seveneight"
- 234567 → "23four5six7"
- 10203 → "1zerotwozero3"

이렇게 숫자의 일부 자릿수가 영단어로 바뀌어졌거나, 혹은 바뀌지 않고 그대로인 문자열 s가 매개변수로 주어집니다. s가 의미하는 원래 숫자를 return 하도록 solution 함수를 완성해주세요.

참고로 각 숫자에 대응되는 영단어는 다음 표와 같습니다.

image.png

## 제한사항

- 1 ≤ s의 길이 ≤ 50
- s가 "zero" 또는 "0"으로 시작하는 경우는 주어지지 않습니다.
- return 값이 1 이상 2,000,000,000 이하의 정수가 되는 올바른 입력만 s로 주어집니다.

입출력 예 #2와 #3과 같이 같은 정답을 가리키는 문자열이 여러 가지가 나올 수 있습니다.

## 첫 번째 방법

1. 각 숫자 대응하는 영단어를 num에 저장.
2. s의 크기만큼 반복하는 반복문으로 s의 문자가 영어일 경우 check에 저장.
3. check에 저장된 문자열이 num에 있을 경우 그 문자열이 num속의 인덱스를 answer에 저장.
4. 숫자일 경우 바로 answer에 저장 후 answer을 int형태로 변환하여 return.

결과 : 성공

```python
def solution(s):
    answer = ''
    check = ''
    num = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
    
    for i in range(len(s)):
        if s[i].isalpha():
            check += s[i]
            if check in num:
                answer += str(num.index(check))
                check = ''
        else:
            answer += s[i]
    return int(answer)
solution("23four5six7")
```
